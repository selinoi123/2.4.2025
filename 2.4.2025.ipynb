```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Selin/Desktop/sport_athletes.csv")
df_copy = df.copy()
df_copy1 = df.copy()
```

```python
df_copy
```

```python
#Question num 1
result = df_copy.groupby('Sport Type').agg({
    'Weight(kg)' : 'min',
    'Height(cm)' : 'max',
    'Wins' : 'max' 
}).reset_index().sort_values(by = 'Sport Type')
display(result)
```

```python
#Question num 1 a b c
result = result.rename(columns={
    'Weight(kg)': 'weight_min',
    'Height(cm)': 'height_max',
    'Wins': 'wins_max'
}).sort_values(by = 'Sport Type')

display(result)
```

```python
#Question num 2
df_sorted = df_copy.sort_values(by = ['Sport Type', 'Height(cm)'], ascending=[True, False])
top2 = df_sorted.groupby('Sport Type').head(2).reset_index()
display(top2[['Sport Type','Height(cm)', 'Athlete Name']])
```
